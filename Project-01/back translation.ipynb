```python
from googletrans import Translator
translator = Translator()

import csv
import re
import numpy as np
import pandas as pd
from io import StringIO
import requests
import matplotlib.pyplot as plt
import demoji



def back_translate(string, src_lang, target_lang):
    translated = translator.translate(string, dest=target_lang).text
    back_translated = translator.translate(translated, dest=src_lang).text
    return back_translated


def back_translated_df(df,target_lang):
    translated_data = pd.DataFrame(columns={'tweet','label'})
    for n in range(0,len(df)):
        try:
            string = df['tweet'].iloc[n]
            src_lang = df['label'].iloc[n]
            translated_string = back_translate(string,src_lang,target_lang)
            translated_data=translated_data.append({'tweet': translated_string, 'label':src_lang}, ignore_index=True)
        except:
            pass
        if n%10==0:
            print("{} tweets translated".format(n))
    return translated_data


```

```python
def load_dataset(url):
    r = requests.get(url)
    data = r.content.decode('utf8')
    df = pd.read_csv(StringIO(data), sep='\t')
    df.columns = ['tweet', 'label']
    return df
```

```python
df = load_dataset(url_test)
df['label'].unique()
```

```python
df=df.drop(df[df['label'] =='en'].index)
df=df.drop(df[df['label'] =='und'].index)
df=df.drop(df[df['label'] =='ja'].index)
```

```python
df['label'].unique()
```

```python
def remove_all_emojis(text):
    dem = demoji.findall(text)
    for item in dem.keys():
        text = text.replace(item, '')
    return text

def  clean_data(df, column):
    df = df.copy(deep=True) # Make deep copy of tweets
    df[column] = df[column].str.lower() # Transform into all lowercase
    
    patterns = []
    retweet_pattern = '^RT'
    patterns.append(retweet_pattern)
    xml_pattern = '&\S+;'
    patterns.append(xml_pattern)
    hashtag_pattern = '#[A-Za-z0-9_]+'
    patterns.append(hashtag_pattern)
    twitter_mention_pattern = '@[A-Za-z0-9_]+'
    patterns.append(twitter_mention_pattern)
    http_pattern = 'http\S+'
    patterns.append(http_pattern)
    www_pattern = 'www\S+'
    patterns.append(www_pattern)
    tab_pattern = '\t'
    patterns.append(tab_pattern)
    punctuation_pattern = '[!"#$%&\\()*+,-./:;<=>?@\[\]^_`\'{}~！？。，、「」]+'
    patterns.append(punctuation_pattern)
    numeric_pattern = '[0-9]+'
    patterns.append(numeric_pattern)
    regex = "|".join(patterns)

    #df[column] = df[column].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem)) 
    df[column] = df[column].apply(lambda elem: re.sub(r"{}".format(regex), "", elem))
    df[column] = df[column].apply(remove_all_emojis)
    
    return df
```

```python
# clean
cleaned_data = clean_data(df, "tweet")
# choose 100 samples
cleaned_data = cleaned_data.sample(20,random_state=None)
cleaned_data.reset_index()
```

```python
# make sure their is no NAN values
cleaned_data=cleaned_data.dropna(axis=0,how='any')
cleaned_data.isnull().values.any()
```
